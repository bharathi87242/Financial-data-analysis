```python
import pandas as pd
import matplotlib.pyplot as plt

forecast_df = pd.read_csv('/content/financial_forecasting_dataset.csv')
technical_df=pd.read_csv('/content/stock_market_dataset.csv')
technical_df.head()
forecast_df.head()
```

```python
print("Forecasting Dataset:")
print(forecast_df.head())

print("\nTechnical Indicators Dataset:")
print(technical_df.head())
```

```python
forecast_df['Date']=pd.to_datetime(forecast_df['Date'])
technical_df['Date']=pd.to_datetime(technical_df['Date'])
```

```python
forecast_df.info()
```

```python
forecast_df.dropna(inplace=True)
```

```python
technical_df.dropna(inplace=True)
```

```python
merge_df = pd.merge(forecast_df, technical_df, left_on=['Date', 'Ticker'], right_on=['Date', 'Stock'])
print("merged dataset:")
print(merge_df.head())
```

```python
# Calculate daily return
merge_df['Daily Return'] = merge_df.groupby('Ticker')['Close_x'].pct_change()

# Normalize macroeconomic indicators
from sklearn.preprocessing import MinMaxScaler
macro_cols = ['GDP (%)', 'Inflation (%)', 'Interest Rate (%)', 'Unemployment (%)']
scaler = MinMaxScaler()
merge_df[macro_cols] = scaler.fit_transform(merge_df[macro_cols])

```

```python
print("Generating plot...")
plt.figure(figsize=(10, 5))
for ticker in merge_df['Ticker'].unique()[:3]:  # Plot first 3 tickers
    subset = merge_df[merge_df['Ticker'] == ticker]
    plt.plot(subset['Date'], subset['Close_x'], label=ticker)

plt.title('Stock Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(merge_df[['Daily Return'] + macro_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Returns and Macroeconomic Indicators')
plt.show()
```

```python
merge_df.to_csv('cleaned_financial_data.csv', index=False)
```
